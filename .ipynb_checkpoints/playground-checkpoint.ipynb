# Remove any baseline from the data

```python
def flatten(curve, lower=620, upper=677, n_lower=100, n_upper=100, debug=False):
    """
    Flattens a curve to make it easier for fitting.
    """
    if debug:
        print("Flattening with:\nlower={0}\nupper={1}\nn_lower={2}\nn_upper={3}\n".format(lower, upper, n_lower, n_upper))
        
    curve = curve.reset_index()
    # check that the region we fit with is ok
    if curve.loc[n_lower]['wavenumber'] > lower:
        print("Warning, lower limit is exceeded by use of n={0}".format(n_lower))
    if curve.loc[len(curve)-1-n_upper]['wavenumber'] < upper:
        print("Warning, upper limit is exceeded by use of n={0}".format(n_upper))
        
    if debug:
        print("curve region is between {0:.3f} and {1:.3f}".format(curve.loc[n_lower]['wavenumber'], curve.loc[len(curve)-1-n_upper]['wavenumber']))
    
    y1 = np.mean(curve[:n_lower]['tau'])
    y2 = np.mean(curve[-1*n_upper:]['tau'])
    x1 = np.mean(curve[:n_lower]['wavenumber'])
    x2 = np.mean(curve[-1*n_upper:]['wavenumber'])
    
    if debug:
        print("y1={0}\ny2={1}\nx1={2}\nx2={3}\n".format(y1, y2, x1, x2))
    
    # compute slope
    m = (y2-y1)/(x2-x1)
    
    if debug:
        print("m={0}\n".format(m))
    
    # create linear function in point-slope form
    X = curve['wavenumber']
    Y = (m*(X-x1) + y1)
    
    flattened = [this_y-this_Y for this_y, this_Y in zip(curve['tau'], Y)]
    # normalize to the same peak as before
    # actually do not do this as it changes the profile!
    #flattened = (flattened/np.max(flattened))*np.max(curve['tau'])
    
    # subtract to get the baseline at 0
    fy1 = np.mean(flattened[:n_upper])
    fy2 = np.mean(flattened[-n_lower:])
    baseline = np.mean([fy1, fy2])
    
    if debug:
        print("fy1={0}\nfy2={1}\nbaseline={2}\n".format(fy1, fy2, baseline))
    
    flattened_subtracted = flattened - baseline
    
    if debug:
        return flattened_subtracted, flattened
    else:
        return flattened_subtracted
```

# Combine data

```python
def add_curves(curves):
    """
    Returns a linear combination of two curves
    """
    for curve in curves:
        # skip combined; we don't add it to itself
        if curve == "combined":
            continue
        
        # the stuff to combine
        this_wavenumber = curves[curve]['data']['wavenumber']
        this_A = curves[curve]['scale']
        this_tau = this_A*curves[curve]['data']['flattened_tau']
        
        interp_tau = np.interp(x=curves['combined']['wavenumber'], xp=this_wavenumber, fp=this_tau)
        
        # if this is the first curve added, just make it combined
        if not any(curves['combined']['tau']):
            curves['combined']['tau'] = interp_tau
        # otherwise, we need to add them
        else:
            curves['combined']['tau'] = [y1 + y2 for y1, y2 in zip(curves['combined']['tau'], interp_tau)]
    return curves
```

```python
curves = {"combined":{"wavenumber":data['wavenumber'], "tau":[]}}

#curves["CO2_vB"] = {"data":CO2_vB, "scale":10}
curves["CO2_E"] = {"data":CO2_E, "scale":0.7}
#curves["mixA"] = {"data":mixA, "scale":8.5}
curves["CO2_H2O"] = {"data":CO2_H2O, "scale":2}
curves["CO2_CH3OH"] = {"data":CO2_CH3OH, "scale":5}
curves["CO2_CO"] = {"data":CO2_CO, "scale":0.0}
curves["CO over CO2"] = {"data":CO_over_CO2, "scale":4}

model_name = ""
for curve in curves:
    if curve == "combined":
        continue
    if model_name != "":
        model_name += " + "
    model_name += str(curves[curve]['scale']) + "*{0}".format(curve)
print("The model name is: " + model_name)

curves = add_curves(curves)
```

### Plot the combined curve and its components

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(10, 5)

ax.errorbar(data['wavenumber'], data['tau'], yerr=data['error_tau'], label="JWST - Yang et al.", linestyle='-', marker='o', markersize=2)

# loop over the components
for curve in curves:
    # ignore the combined curve
    if curve == "combined":
        continue
    # plot the component
    scale = curves[curve]['scale']
    ax.plot(curves[curve]['data']['wavenumber'], scale*curves[curve]['data']['flattened_tau'],
            label="{0}*{1}".format(scale, curve))
    
# plot the combined curve
ax.plot(curves['combined']['wavenumber'], curves['combined']['tau'], label="combined", color="xkcd:black", linewidth=3, alpha=1)

ax.set_xlim(700, 600)
ax.set_xlabel("wavenumber (1/cm)")
ax.set_ylabel("Optical Depth");
ax.legend(fontsize=11)
ax.set_title(model_name)
plt.savefig("./{0}.jpg".format(model_name.replace(" ", "")), bbox_inches="tight")
```

# Calculate Fit
Evaluate the goodness of fit by eye, and with the $\chi^2$ parameter

```python
def evaluate_fit(observations, fit_curves, model_name, save_model=True):
    """
    
    """
    
    # calculate chi^2
    chisquare, p = stats.chisquare(observations['tau']/np.sum(observations['tau']), fit_curves['combined']['tau']/np.sum(fit_curves['combined']['tau']))
    
    # print stats
    print("Model: {0} \nChi Square: {1}\np-value: {2}".format(model_name, chisquare, p))
    
    # save stats to csv for future reference
    if save_model:
        df = pd.read_csv("fit_stats.csv")
        df.loc[len(df)] = {"chisquare":chisquare, "p-value":p, "model":model_name}
        df.to_csv("fit_stats.csv", index=False)
```
